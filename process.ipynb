```python
import duckdb

# 连接 DuckDB 数据库
db_path = "data/data.db"
con = duckdb.connect(database=db_path)
```

```python
# 获取数据库中所有表名
tables = con.sql("SHOW TABLES").fetchall()
print(f"Tables in {db_path}: {[table[0] for table in tables]}")
```

```python
df = con.sql(f"SELECT * FROM ctr_data_1M")
```
